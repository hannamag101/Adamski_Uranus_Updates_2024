# Let's use Monte Carlo simulation to calculate pi

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1) Initialize Variables - Inputs
n_simulations = int(1e6) 
n_points_circ = 0
n_points_sq = 0
```

```python
# 2) Choose a Probability Distribution 
# X ~ Unif(-1,1)
# Y ~ Unif(-1,1)

# 3) Generate Input by sampling from Uniform Distribution
l_xs = []
l_ys = []

# loop n times (# of simulations)
for i in range(n_simulations):
    x = np.random.uniform(-1, 1)
    l_xs.append(x)

    y = np.random.uniform(-1, 1)
    l_ys.append(y)

```

```python
# Check whether each point falls within the circle or not
for i in range(n_simulations):
    # Calculate the distance between the point & origin
    dist_from_origin = np.sqrt(l_xs[i]**2 + l_ys[i]**2)

    if dist_from_origin <= 1:
        n_points_circ += 1
    
    n_points_sq += 1
```

```python
(n_points_circ / n_points_sq) * 4
```

```python
# Check whether each point falls within the circle or not - ONLY 1000 INPUTS + Visualization
fig, ax = plt.subplots(figsize = (8,8))
for i in range(n_simulations):
    # Calculate the distance between the point & origin
    dist_from_origin = np.sqrt(l_xs[i]**2 + l_ys[i]**2)

    if dist_from_origin <= 1:
        n_points_circ += 1
        ax.scatter(l_xs[i], l_ys[i], color = 'red')
    else:
        ax.scatter(l_xs[i], l_ys[i], color = 'blue')
    
    n_points_sq += 1
```
